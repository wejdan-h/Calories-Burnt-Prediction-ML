#Split dataset into training (70%) and testing (30%) sets.

We gave the X all the columns except the calories Which represent the input and the Y the calories, which is our target. We divided the test into 70% training and 30% testing.

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['Calories'], axis=1)
Y = df['Calories']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=5)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Normalize or scale numerical data:

StandardScaler


We stored the standard scaler in the variable scaler. We used scaler fit and transform on the numeric X_training data and only used transform on the test data (Note: We did not use fit on the test data because we do not want to give prior knowledge of the test data).

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train[['Age','Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp']] = scaler.fit_transform(X_train[['Age','Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp']])
```

```python
X_test[['Age','Height','Weight', 'Duration', 'Heart_Rate','Body_Temp']] = scaler.transform(X_test[['Age','Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp']])
```

```python
X_train.head()
```

```python
X_test.head()
```

#Build the model:

Supervised learning model: Regression: Linear Regression


```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

Train the model using the training data

```python
reg.fit(X_train, Y_train)
```

Make predictions on the test data

```python
lr_pred = reg.predict(X_test)
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor(max_depth=5, random_state=4)
```

Train the model using the training data

```python
regressor.fit(X_train, Y_train)
```

Make predictions on the test data

```python
dt_pred = regressor.predict(X_test)
```

#Model Evaluation

We imported these three functions from the sklearn.metrics library: mean_absolute_error, mean_squared_error, and r2_score.
(mean_absolute_error: measures the average error between the true and expected values ​​and gives all errors equal weight)
(mean_squared_error: measures the mean squared of errors and focuses on large errors)
(r2_score: measures how well the independent data explains the dependent data)

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

Mean Squared Error(MSE)

```python
mean_squared_error(Y_test, lr_pred)
```

```python
mean_squared_error(Y_test, dt_pred)
```

Mean Absolute Error(MAE)

```python
mean_absolute_error(Y_test, lr_pred)
```

```python
mean_absolute_error(Y_test, dt_pred)
```

$R^2$ score

```python
r2_score(Y_test, lr_pred)
```

```python
r2_score(Y_test, dt_pred)
```
